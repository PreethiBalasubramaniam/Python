```python
import random

f = open("tictactoe.txt", "a")
def drawPattern(Pattern):
    #It prints out the " Tic-Tac-Toe game Pattern" and its list of string
    print('   |   |')
    print(' ' + Pattern[1] + ' | ' + Pattern[2] + ' | ' + Pattern[3])  #1st row and column
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + Pattern[4] + ' | ' + Pattern[5] + ' | ' + Pattern[6]) #2nd row and column
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + Pattern[7] + ' | ' + Pattern[8] + ' | ' + Pattern[9]) #3rd row and column
    print('   |   |')
 
def inputPlayerchoice():
    # let us print the X or O which we want to play
    # Returns a list with the player's choice  and the computer's choice.
    choice = ''
    while not (choice == 'X' or choice == 'O'):
        print('Do you want to be X or O?') #Player's choice
        choice = input().upper()
   
    if choice == 'X':
        return ['X', 'O']      # the first element is the player's choice,followed by the second is the computer's choice.
    else:
        return ['O', 'X']
 
def whoGoesFirst():
    
    if random.randint(0, 1) == 0:
        return 'computer'    # Randomly it choose the player who goes first.
    else:
        return 'player'
 
def playAgain():
    print('Do you want to play again? (yes or no)')
    return input().lower().startswith('y')    # It returns "Yes" if the player wants to play again, otherwise it ends.
 
def makeMove(Pattern, choice, move):
    Pattern[move] = choice #Game move accordingly
 
def isWinner(bo, ch):
    # Given a Pattern and a player's choice, which returns True if that player has won.
    # Used bo instead of Pattern and ch instead of choice so that everytime don't have to type as much.
    return ((bo[7] == ch and bo[8] == ch and bo[9] == ch) or # across the top
    (bo[4] == ch and bo[5] == ch and bo[6] == ch) or # across the middle
    (bo[1] == ch and bo[2] == ch and bo[3] == ch) or # across the bottom
    (bo[7] == ch and bo[4] == ch and bo[1] == ch) or # down the left side
    (bo[8] == ch and bo[5] == ch and bo[2] == ch) or # down the middle
    (bo[9] == ch and bo[6] == ch and bo[3] == ch) or # down the right side
    (bo[7] == ch and bo[5] == ch and bo[3] == ch) or # diagonal
    (bo[9] == ch and bo[5] == ch and bo[1] == ch)) # diagonal
 
def getPatternCopy(Pattern):  
    dupliPattern = []
 
    for i in Pattern:    
        dupliPattern.append(i)  # Make a duplicate of the Pattern list and return it the duplicate.
 
    return dupliPattern
 
def isSpaceFree(Pattern, move):
    
    return Pattern[move] == ' '   # Return True if the passed move is free on the passed Pattern.
 
def getPlayerMove(Pattern):
    
    move = ' '
    while move not in '1 2 3 4 5 6 7 8 9'.split() or not isSpaceFree(Pattern, int(move)): #guess the next move of the player
        print('What is your next move? (1-9)')   
        move = input()     # Let the player type in their turn.
    return int(move)
 
def chooseRandomMoveFromList(Pattern, movesList): 
    
    expectedMoves = []
    for i in movesList:
        if isSpaceFree(Pattern, i):
            expectedMoves.append(i)   # Returns a valid move from the list on the Pattern.
 
    if len(expectedMoves) != 0:
        return random.choice(expectedMoves)
    else:
        return None              # Returns None(Nothing) if there is no valid move.
 
def getComputerMove(Pattern, computerchoice):
   
    if computerchoice == 'X':    # A Pattern and the computer's choice, determine where to move and return that move.
        playerchoice = 'O'
    else:
        playerchoice = 'X'
 
    # Algorithm for our Tic Tac Toe AI:  
    for i in range(1, 10):
        copy = getPatternCopy(Pattern)
        if isSpaceFree(copy, i):
            makeMove(copy, computerchoice, i)   
            if isWinner(copy, computerchoice):   # Check if we can win in the next move
                return i
 
    for i in range(1, 10):
        copy = getPatternCopy(Pattern)
        if isSpaceFree(copy, i):
            makeMove(copy, playerchoice, i)   # Check if the player could win on their next move, and thinks to block them.
            if isWinner(copy, playerchoice):
                return i
    
    move = chooseRandomMoveFromList(Pattern, [1, 3, 7, 9])
    if move != None:    # Try to take one of the corners, if they are free.
        return move
 
    if isSpaceFree(Pattern, 5):    # Try to capture the center, if it is free.
        return 5
   
    return chooseRandomMoveFromList(Pattern, [2, 4, 6, 8])   # Move to one of it's sides.
 
def isPatternFull(Pattern):
    
    for i in range(1, 10):
        if isSpaceFree(Pattern, i):
            return False     # Return True if every space on the Pattern has been Occupied. Otherwise return False.
    return True

 
print('Welcome to Tic-Tac-Toe!')
 
while True:
    
    thePattern = [' '] * 10
    playerchoice, computerchoice = inputPlayerchoice()
    turn = whoGoesFirst()                              # Reset the Pattern
    print('The ' + turn + ' will go first.')    #randomly choose the player
    gameIsPlaying = True
 
    while gameIsPlaying:
        if turn == 'player':
            drawPattern(thePattern)
            move = getPlayerMove(thePattern)    # Player's turn.
            makeMove(thePattern, playerchoice, move)
 
            if isWinner(thePattern, playerchoice):
                drawPattern(thePattern)
                print('Hooray! You have won the game!')  #If player wins the following message will be displayed!
                gameIsPlaying = False
            else:
                if isPatternFull(thePattern):
                    drawPattern(thePattern)
                    print('The game is a tie!')   #If none of them wins
                    break
                else:
                    turn = 'computer'
 
        else:
            # Computer's turn.
            move = getComputerMove(thePattern, computerchoice)
            makeMove(thePattern, computerchoice, move)
 
            if isWinner(thePattern, computerchoice):
                drawPattern(thePattern)
                print('The computer has beaten you! You lose.')
                gameIsPlaying = False
            else:
                if isPatternFull(thePattern):
                    drawPattern(thePattern)
                    print('The game is a tie!')
                    break
                else:
                    turn = 'player'
 
    if not playAgain():   
        break

            


f.write(str(thePattern))
f.close()
```
